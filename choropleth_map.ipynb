```python
import pandas as pd
import altair as alt
import numpy as np
import geopandas as gpd
from vega_datasets import data  
import plotly.express as px
```

```python
stateDF=pd.read_csv('assets/wb_state_temps.csv', header=1)
stateDF.columns
```

```python
stateDF = stateDF.rename(columns = {stateDF.columns[0]:'Year'})
stateDF2=stateDF.drop(['United States'], axis=1)
# stateDF.iloc[::10,:]
display(stateDF2)
```

```python
state_dict={'year':[], 'state':[],'temperature':[]}
allstates = list(stateDF2.columns)
allstates.remove('Year')
# print(allstates)

allyears=stateDF2.Year.unique()

for year in allyears:
    for state in allstates:
        state_dict['year']+=[year]
        state_dict['state']+=[state]
        state_dict['temperature']+=[stateDF2.loc[(stateDF2.Year==year)][state].iloc[0]]
plotdf=pd.DataFrame(state_dict)
display(plotdf)
```

```python
def funcdelta(year1,year2,df):
    state_dict={'state':[],'temperature':[]}
    for state in df.state.unique():
        state_dict['state']+=[state]
        temp2=df.loc[(df.state==state) & (df.year==year2)].temperature.iloc[0]
        temp1=df.loc[(df.state==state) & (df.year==year1)].temperature.iloc[0]
        state_dict['temperature']+= [temp2 - temp1]
    return pd.DataFrame(state_dict)
```

```python
display(funcdelta(1901,2020,plotdf))
```

```python
ansi = pd.read_csv('https://www2.census.gov/geo/docs/reference/state.txt', sep='|')
ansi.columns = ['id', 'abbr', 'state', 'statens']
ansi = ansi[['id', 'abbr', 'state']]
def mergeWansi(df):
    return pd.merge(df, ansi, how='left', on='state')
# plotdf = pd.merge(plotdf, ansi, how='left', on='state')
plotdf=mergeWansi(plotdf)
display(plotdf)
```

```python
display(plotdf)
```

<b>This was the original plot I created. You can change the year under transform_lookup to see the temps in different years. It's not as beneficial because from year to year, it doesn't change enough to see anything significant since they're all going based off of 0 degrees celsius. </b>

```python
states = alt.topo_feature(data.us_10m.url, feature='states')

base = alt.Chart(states).mark_geoshape(fill='lightgray', stroke='black', strokeWidth=0.5)

chart = alt.Chart(states).mark_geoshape(stroke='black').encode(
    color='temperature:Q',
    tooltip=['temperature:Q']
).transform_lookup(
    lookup='id',
    from_=alt.LookupData(plotdf.loc[plotdf.year==1901], 'id', ['temperature'])
).properties(
    width=500,
    height=300
).project('albersUsa')


base + chart

```

<b>This is the final plot. It's the change in temperatures over time. I did from 1901 to 2020, but you can always change them if you want to play around with it under the funcdelta function. This shows which states have gotten hotter over the last 100 years and which haven't gotten as hot. We can always play around with the color scheme too if we want- it was blue like above but I thought the red was better with redder being hotter.</b>

```python
base1 = alt.Chart(states).mark_geoshape(fill='lightgray', stroke='black', strokeWidth=0.5)

deltdf=mergeWansi(funcdelta(1901,2020,plotdf))
display(deltdf)
chart1 = alt.Chart(states).mark_geoshape(stroke='black').encode(
    color=alt.Color('temperature:Q',scale=alt.Scale(scheme='orangered')),
    tooltip=['temperature:Q']
).transform_lookup(
    lookup='id',
    from_=alt.LookupData(deltdf, 'id', ['temperature'])
).properties(
    width=500,
    height=300
).project('albersUsa')

base1 + chart1
```

```python
tempDF=pd.read_csv('assets/state_temps.csv')
display(tempDF)
```

```python
tempDF['Reliable']=tempDF['Crude Rate'].apply(lambda x: 0 if ('Unreliable' in x) else 1)
display(tempDF)
```

```python
tempDF['CrudeFloat']=tempDF['Crude Rate'].apply(lambda x: float(x.split(' ')[0]) if ('Unreliable' in x) else float(x))
display(tempDF)
```

```python
tempDF['DPM']=tempDF.Deaths/tempDF.Population*1e6
tempDF['state']=tempDF['State']
display(tempDF)
```

```python
tempDF=tempDF[['state','Deaths','Reliable','DPM']]
# display(tempDF)
# tempDF['stateDeaths']=tempDF.groupby(['State'],as_index=False)#['Deaths'].sum()
cleanDF=pd.DataFrame(columns=tempDF.columns)
statelist=tempDF.state.unique()
for state in statelist:
    if len(tempDF.loc[tempDF.state==state])==1:
        cleanDF=pd.concat([cleanDF,tempDF.loc[tempDF.state==state]])
    else:
        tempstateDF=tempDF.loc[tempDF.state==state].reset_index(drop=True,inplace=False)
        # display(tempstateDF)
        stateDict={}
        stateDict['state']=[state]
        stateDict['Deaths']=[tempstateDF.Deaths.sum()]
        stateDict['Reliable']=[np.product(np.array(tempstateDF.Reliable))]
        stateDict['DPM']=[tempstateDF.DPM.mean()]
        cleanDF=pd.concat([cleanDF,pd.DataFrame(stateDict)])
        # tempstateDF['Deaths']=tempstateDF.groupby(['State'],as_index=False).Deaths.sum()
        # tempstateDF['Deaths']=tempstateDF['State'].apply(lambda x: if x==x tempstateDF['Deaths'].sum())
        # stateDict['DPM']=stateDict.groupby['State'],as_index=False)['DPM'].mean()
        # cleanDF=pd.concat([cleanDF,tempstateDF.loc[[0]]])
cleanDF.DPM=np.round_(cleanDF.DPM,decimals=2)
cleanDF.reset_index(drop=True, inplace=True)
        
display(cleanDF)
```

```python
wansi_cleanDF=mergeWansi(cleanDF)
display(wansi_cleanDF)
```
