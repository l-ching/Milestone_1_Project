```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
df = pd.read_csv('heat-waves_fig-1.csv', header=6)
```

```python
df.columns
```

```python
fig, ax3 = plt.subplots()
ax3.bar('Decade','Frequency (average number of heat waves per year)', data=df,color='#B85741', alpha=0.6)
ax2 = ax3.twinx()   
ax2.set_ylabel('Deaths (per million)', color = 'navy') 
ax3.set_xlabel('Decade')
ax2.plot('Decade','Average heat-related deaths',data=df, color='navy')
ax3.set_ylabel('Frequency', color='#B85741')
ax3.set_title('Relationship Between Frequency and Heat Related Deaths')
```

```python
fig, ax3 = plt.subplots()
ax3.bar('Decade','Intensity (average temperature above local threshold during heat waves)', data=df,color='#009384', alpha=0.6)
ax2 = ax3.twinx()   
ax2.set_ylabel('Deaths (per million)', color = 'navy') 
ax3.set_xlabel('Decade')
ax2.plot('Decade','Average heat-related deaths',data=df, color='navy')
ax3.set_ylabel('Intensity', color='#009384')
ax3.set_title('Relationship Between Intensity and Heat Related Deaths')
```
