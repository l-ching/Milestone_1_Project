```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
df = pd.read_csv('heat-deaths_fig-1.csv', header=6)
```

```python
df.columns
```

```python
deaths = df['Underlying cause of death (all year)']
scaled_deaths = [value * 500 for value in deaths]
```

```python
fig,ax1=plt.subplots()
ax1.scatter('Year', 
            'Heat wave index', 
            alpha=0.5,
            s= scaled_deaths,
            data=df,
           color='#009384')

ax1.set_title('Heat Related Mortality Over Time')
ax1.set_xlabel('Year')
ax1.set_ylabel('Heat Wave Index', color='#009384')
```
